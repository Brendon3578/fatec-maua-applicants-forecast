```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# Novos dados fornecidos

new_data = {
    'Ano': [
        2007, 2007, 2008, 2008, 2009, 2009, 2010, 2010,
        2011, 2011, 2012, 2012, 2012, 2013, 2013, 2013,
        2014, 2014, 2014, 2015, 2015, 2015, 2016, 2016, 2016,
        2017, 2017, 2017, 2018, 2018, 2018, 2019, 2019, 2019,
        2020, 2020, 2020, 2021, 2021, 2021, 2022, 2022, 2022,
        2023, 2023, 2023, 2024, 2024, 2024
    ],
    'NomeCurso': [
        'Informática para Negócios (diurno)', 'Logística  (diurno)',
        'Informática para Negócios (diurno)', 'Logística  (diurno)',
        'Informática para Negócios (diurno)', 'Logística  (diurno)',
        'Informática para Negócios (diurno)', 'Logística  (diurno)',
        'Informática para Negócios (diurno)', 'Logística  (diurno)',
        'Fabricação Mecânica (diurno)', 'Informática para Negócios (diurno)', 'Logística  (diurno)',
        'Fabricação Mecânica (diurno)', 'Informática para Negócios (diurno)', 'Logística  (diurno)',
        'Fabricação Mecânica (diurno)', 'Informática para Negócios (diurno)', 'Logística  (diurno)',
        'Fabricação Mecânica (diurno)', 'Informática para Negócios (diurno)', 'Logística  (diurno)',
        'Fabricação Mecânica (diurno)', 'Informática para Negócios (diurno)', 'Logística  (diurno)',
        'Fabricação Mecânica (diurno)', 'Informática para Negócios (diurno)', 'Logística  (diurno)',
        'Fabricação Mecânica (diurno)', 'Informática para Negócios (diurno)', 'Logística  (diurno)',
        'Fabricação Mecânica (diurno)', 'Informática para Negócios (diurno)', 'Logística  (diurno)',
        'Fabricação Mecânica (diurno)', 'Informática para Negócios (diurno)', 'Logística  (diurno)',
        'Fabricação Mecânica (diurno)', 'Informática para Negócios (diurno)', 'Logística  (diurno)',
        'Fabricação Mecânica (diurno)', 'Informática para Negócios (diurno)', 'Logística  (diurno)',
        'Fabricação Mecânica (diurno)', 'Informática para Negócios (diurno)', 'Logística  (diurno)',
        'Fabricação Mecânica (diurno)', 'Informática para Negócios (diurno)', 'Logística  (diurno)'
    ],
    'Inscritos': [
        112, 0,
        65,0,
        86, 149,
        60, 221,
        74, 216,
        103, 41, 112,
        166, 77, 130,
        109, 45, 109,
        92, 60, 109,
        81, 90, 98,
        94, 87, 110,
        92, 92, 133,
        81, 104, 123,
        95, 152, 213,
        82, 101, 140,
        54, 57, 61,
        52, 60, 67,
        48, 42, 44
    ]
}



df = pd.DataFrame(new_data)

# Obter a lista de cursos únicos
unique_courses = df['NomeCurso'].unique()

# Definir uma paleta de cores para os gráficos
colors = plt.cm.get_cmap('tab10', len(unique_courses))

# Inicia a figura para o gráfico combinado
plt.figure(figsize=(19, 9)) # Aumenta o tamanho do gráfico para melhor visualização de múltiplas linhas e rótulos

# Lista para armazenar os DataFrames de cada curso para a tabela final
all_course_data_for_table = []

# Iterar sobre cada curso para realizar a previsão e plotar
for i, course_name in enumerate(unique_courses):
    print(f"\n--- Processando o curso: {course_name} ---")

    # Filtrar os dados para o curso atual
    df_course = df[df['NomeCurso'] == course_name].copy()

    # Preparar os dados para o Prophet
    df_prophet = df_course[['Ano', 'Inscritos']].copy()
    df_prophet.columns = ['ds', 'y']
    df_prophet['ds'] = pd.to_datetime(df_prophet['ds'], format='%Y')

    # Inicializar e treinar o modelo Prophet
    modelo = Prophet(
        yearly_seasonality=True,  # Considera sazonalidade anual
        changepoint_prior_scale=0.05 # Ajuste para flexibilidade do modelo
    )
    modelo.fit(df_prophet)

    # Criar um dataframe para as previsões futuras até 2028
    # A previsão será para 4 anos após o último ano histórico presente nos dados (2024)
    # Isso inclui 2025, 2026, 2027 e 2028
    future = modelo.make_future_dataframe(periods=4, freq='Y')

    # Fazer as previsões
    forecast = modelo.predict(future)

    # Aplicar a regra de não ter valores negativos para yhat, yhat_lower e yhat_upper
    forecast['yhat'] = forecast['yhat'].apply(lambda x: max(0, x))
    forecast['yhat_lower'] = forecast['yhat_lower'].apply(lambda x: max(0, x))
    forecast['yhat_upper'] = forecast['yhat_upper'].apply(lambda x: max(0, x))


    # Exibir as previsões para os anos futuros (2025-2028)
    print(f"Previsão para {course_name} (2025-2028):")
    # Filtrar apenas as previsões para os anos futuros (a partir de 2025)
    forecast_future_years = forecast[forecast['ds'].dt.year > df_prophet['ds'].max().year].copy()
    forecast_future_years['ds'] = forecast_future_years['ds'].apply(lambda x: pd.to_datetime(f"{x.year}-01-01"))
    # A impressão ainda incluirá os intervalos de confiança para referência no console.
    print(forecast_future_years[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(4).round(0).astype(int))

    # Plotar dados históricos para o curso atual no gráfico combinado
    plt.plot(df_prophet['ds'], df_prophet['y'], marker='o', linestyle='-', color=colors(i), label=f'{course_name} - Histórico')

    # Adicionar rótulos aos dados históricos
    for _, row in df_prophet.iterrows():
        plt.text(row['ds'], row['y'], str(int(row['y'])), fontsize=8, ha='center', va='bottom', color=colors(i))

    # Plotar a previsão (apenas para os anos futuros) no gráfico combinado
    plt.plot(forecast_future_years['ds'], forecast_future_years['yhat'], marker='o', linestyle='--', color=colors(i), label=f'{course_name} - Previsão')

    # Adicionar rótulos aos pontos de previsão (apenas para os anos futuros)
    for _, row in forecast_future_years.iterrows():
        plt.text(row['ds'], row['yhat'] + 5, str(int(round(row['yhat']))), fontsize=8, ha='center', va='bottom', color=colors(i))


    # Preparar dados históricos para a tabela
    df_historical_for_table = df_prophet.copy()
    df_historical_for_table['Tipo'] = 'Histórico'
    df_historical_for_table = df_historical_for_table[['ds', 'y', 'Tipo']]
    df_historical_for_table.columns = ['Ano', 'Inscritos', 'Tipo']
    df_historical_for_table['Ano'] = df_historical_for_table['Ano'].dt.year

    # Preparar dados de previsão para a tabela
    df_forecast_for_table = forecast_future_years.copy()
    df_forecast_for_table['Tipo'] = 'Previsão'
    df_forecast_for_table = df_forecast_for_table[['ds', 'yhat', 'Tipo']]
    df_forecast_for_table.columns = ['Ano', 'Inscritos', 'Tipo']
    df_forecast_for_table['Ano'] = df_forecast_for_table['Ano'].dt.year
    df_forecast_for_table['Inscritos'] = df_forecast_for_table['Inscritos'].round(0).astype(int) # Arredonda para inteiros

    # Combinar histórico e previsão para o curso atual e adicionar à lista
    combined_course_df = pd.concat([df_historical_for_table, df_forecast_for_table])
    combined_course_df['NomeCurso'] = course_name
    all_course_data_for_table.append(combined_course_df)

# Ajustar os limites do eixo Y (para o gráfico combinado)
plt.ylim(min(df['Inscritos'].min(), forecast['yhat_lower'].min(), 0) * 0.9,
         max(df['Inscritos'].max(), forecast['yhat_upper'].max()) * 1.1)

# Configurar o eixo x para exibir todos os anos relevantes
all_years_in_plot = pd.concat([df['Ano'].apply(lambda x: pd.to_datetime(str(x), format='%Y')), future['ds']]).dt.year.unique()
all_years_in_plot.sort()

plt.xticks([pd.to_datetime(str(year), format='%Y') for year in all_years_in_plot],
           [str(year) for year in all_years_in_plot],
           rotation=45, ha='right')

plt.xlabel('Ano')
plt.ylabel('Número de Inscritos')
# Título do gráfico combinado atualizado
plt.title('Vestibular de 2º Semestre (diurno) - Inscritos por Curso (Histórico e Projeção) com Prophet')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show() # Exibe o gráfico combinado

# --- Impressão da Tabela Completa de Dados Históricos e Previsões (Formato de Texto) ---
# Concatenar todos os dados dos cursos em um único DataFrame para a tabela final
final_table_df = pd.concat(all_course_data_for_table)
final_table_df = final_table_df.sort_values(by=['Ano', 'NomeCurso']).reset_index(drop=True)

# Pivotar o DataFrame para o formato desejado (Ano como índice, NomeCurso como colunas)
pivot_table_df = final_table_df.pivot_table(index='Ano', columns='NomeCurso', values='Inscritos', aggfunc='first')

# Título da tabela atualizado
print("\n--- Vestibular 2º Semestre - Inscritos por Curso (Histórico e Projeção) com Prophet ---")
# Imprimir a tabela em formato de texto.
print(pivot_table_df.to_string())
```

```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# Novos dados fornecidos
new_data = {
    'Ano': [
        2015, 2016, 2017, 2018, 2019,
        2020, 2021, 2022, 2023, 2024
    ],
    'NomeCurso': [
        'Gestão Empresarial', 'Gestão Empresarial', 'Gestão Empresarial', 'Gestão Empresarial', 'Gestão Empresarial',
        'Gestão Empresarial', 'Gestão Empresarial', 'Gestão Empresarial', 'Gestão Empresarial', 'Gestão Empresarial'
    ],
    'Inscritos': [
        187, 193, 187, 0, 0,
        0, 422, 159, 173, 170
    ]
}

df = pd.DataFrame(new_data)

# Obter a lista de cursos únicos
unique_courses = df['NomeCurso'].unique()

# Definir uma paleta de cores para os gráficos
colors = plt.cm.get_cmap('tab10', len(unique_courses))

# Inicia a figura para o gráfico combinado
plt.figure(figsize=(19, 9)) # Aumenta o tamanho do gráfico para melhor visualização de múltiplas linhas e rótulos

# Lista para armazenar os DataFrames de cada curso para a tabela final
all_course_data_for_table = []

# Iterar sobre cada curso para realizar a previsão e plotar
for i, course_name in enumerate(unique_courses):
    print(f"\n--- Processando o curso: {course_name} ---")

    # Filtrar os dados para o curso atual
    df_course = df[df['NomeCurso'] == course_name].copy()

    # Preparar os dados para o Prophet
    df_prophet = df_course[['Ano', 'Inscritos']].copy()
    df_prophet.columns = ['ds', 'y']
    df_prophet['ds'] = pd.to_datetime(df_prophet['ds'], format='%Y')

    # Inicializar e treinar o modelo Prophet
    modelo = Prophet(
        yearly_seasonality=True,  # Considera sazonalidade anual
        changepoint_prior_scale=0.05 # Ajuste para flexibilidade do modelo
    )
    modelo.fit(df_prophet)

    # Criar um dataframe para as previsões futuras até 2028
    # A previsão será para 4 anos após o último ano histórico presente nos dados (2024)
    # Isso inclui 2025, 2026, 2027 e 2028
    future = modelo.make_future_dataframe(periods=4, freq='Y')

    # Fazer as previsões
    forecast = modelo.predict(future)

    # Aplicar a regra de não ter valores negativos para yhat, yhat_lower e yhat_upper
    forecast['yhat'] = forecast['yhat'].apply(lambda x: max(0, x))
    forecast['yhat_lower'] = forecast['yhat_lower'].apply(lambda x: max(0, x))
    forecast['yhat_upper'] = forecast['yhat_upper'].apply(lambda x: max(0, x))


    # Exibir as previsões para os anos futuros (2025-2028)
    print(f"Previsão para {course_name} (2025-2028):")
    # Filtrar apenas as previsões para os anos futuros (a partir de 2025)
    forecast_future_years = forecast[forecast['ds'].dt.year > df_prophet['ds'].max().year].copy()
    forecast_future_years['ds'] = forecast_future_years['ds'].apply(lambda x: pd.to_datetime(f"{x.year}-01-01"))
    # A impressão ainda incluirá os intervalos de confiança para referência no console.
    print(forecast_future_years[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(4).round(0).astype(int))

    # Plotar dados históricos para o curso atual no gráfico combinado
    plt.plot(df_prophet['ds'], df_prophet['y'], marker='o', linestyle='-', color=colors(i), label=f'{course_name} - Histórico')

    # Adicionar rótulos aos dados históricos
    for _, row in df_prophet.iterrows():
        plt.text(row['ds'], row['y'], str(int(row['y'])), fontsize=8, ha='center', va='bottom', color=colors(i))

    # Plotar a previsão (apenas para os anos futuros) no gráfico combinado
    plt.plot(forecast_future_years['ds'], forecast_future_years['yhat'], marker='o', linestyle='--', color=colors(i), label=f'{course_name} - Previsão')

    # Adicionar rótulos aos pontos de previsão (apenas para os anos futuros)
    for _, row in forecast_future_years.iterrows():
        plt.text(row['ds'], row['yhat'] + 5, str(int(round(row['yhat']))), fontsize=8, ha='center', va='bottom', color=colors(i))


    # Preparar dados históricos para a tabela
    df_historical_for_table = df_prophet.copy()
    df_historical_for_table['Tipo'] = 'Histórico'
    df_historical_for_table = df_historical_for_table[['ds', 'y', 'Tipo']]
    df_historical_for_table.columns = ['Ano', 'Inscritos', 'Tipo']
    df_historical_for_table['Ano'] = df_historical_for_table['Ano'].dt.year

    # Preparar dados de previsão para a tabela
    df_forecast_for_table = forecast_future_years.copy()
    df_forecast_for_table['Tipo'] = 'Previsão'
    df_forecast_for_table = df_forecast_for_table[['ds', 'yhat', 'Tipo']]
    df_forecast_for_table.columns = ['Ano', 'Inscritos', 'Tipo']
    df_forecast_for_table['Ano'] = df_forecast_for_table['Ano'].dt.year
    df_forecast_for_table['Inscritos'] = df_forecast_for_table['Inscritos'].round(0).astype(int) # Arredonda para inteiros

    # Combinar histórico e previsão para o curso atual e adicionar à lista
    combined_course_df = pd.concat([df_historical_for_table, df_forecast_for_table])
    combined_course_df['NomeCurso'] = course_name
    all_course_data_for_table.append(combined_course_df)

# Ajustar os limites do eixo Y (para o gráfico combinado)
plt.ylim(min(df['Inscritos'].min(), forecast['yhat_lower'].min(), 0) * 0.9,
         max(df['Inscritos'].max(), forecast['yhat_upper'].max()) * 1.1)

# Configurar o eixo x para exibir todos os anos relevantes
all_years_in_plot = pd.concat([df['Ano'].apply(lambda x: pd.to_datetime(str(x), format='%Y')), future['ds']]).dt.year.unique()
all_years_in_plot.sort()

plt.xticks([pd.to_datetime(str(year), format='%Y') for year in all_years_in_plot],
           [str(year) for year in all_years_in_plot],
           rotation=45, ha='right')

plt.xlabel('Ano')
plt.ylabel('Número de Inscritos')
# Título do gráfico combinado atualizado
plt.title('Vestibular de 2º Semestre - Inscritos por Curso (Histórico e Projeção) com Prophet')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show() # Exibe o gráfico combinado

# --- Impressão da Tabela Completa de Dados Históricos e Previsões (Formato de Texto) ---
# Concatenar todos os dados dos cursos em um único DataFrame para a tabela final
final_table_df = pd.concat(all_course_data_for_table)
final_table_df = final_table_df.sort_values(by=['Ano', 'NomeCurso']).reset_index(drop=True)

# Pivotar o DataFrame para o formato desejado (Ano como índice, NomeCurso como colunas)
pivot_table_df = final_table_df.pivot_table(index='Ano', columns='NomeCurso', values='Inscritos', aggfunc='first')

# Título da tabela atualizado
print("\n--- Vestibular 2º Semestre - Inscritos por Curso (Histórico e Projeção) com Prophet ---")
# Imprimir a tabela em formato de texto.
print(pivot_table_df.to_string())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados fornecidos
new_data = {
    'Ano': [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
    'NomeCurso': ['Gestão Empresarial'] * 10,
    'Inscritos': [187, 193, 187, 0, 0, 0, 422, 159, 173, 170]
}

df = pd.DataFrame(new_data)

# Lista de cursos únicos
unique_courses = df['NomeCurso'].unique()

# Paleta de cores
colors = plt.cm.get_cmap('tab10', len(unique_courses))

# Início do gráfico
plt.figure(figsize=(19, 9))

# Lista para armazenar dados para a tabela final
all_course_data_for_table = []

for i, course_name in enumerate(unique_courses):
    print(f"\n--- Processando o curso: {course_name} ---")

    # Filtra dados do curso
    df_course = df[df['NomeCurso'] == course_name].copy()
    df_course['ds'] = pd.to_datetime(df_course['Ano'], format='%Y')
    df_course['y'] = df_course['Inscritos']

    # Previsões manuais
    manual_forecast = pd.DataFrame({
        'Ano': [2025, 2026, 2027],
        'Inscritos': [177, 187, 192],
        'Tipo': ['Previsão'] * 3,
        'NomeCurso': [course_name] * 3
    })

    # Dados históricos
    historical = df_course[['Ano', 'Inscritos']].copy()
    historical['Tipo'] = 'Histórico'
    historical['NomeCurso'] = course_name

    # Junta histórico e previsão
    combined_df = pd.concat([historical, manual_forecast])
    all_course_data_for_table.append(combined_df)

    # Plot histórico
    plt.plot(pd.to_datetime(historical['Ano'], format='%Y'),
             historical['Inscritos'], marker='o', linestyle='-',
             color=colors(i), label=f'{course_name} - Histórico')

    for _, row in historical.iterrows():
        plt.text(pd.to_datetime(str(row['Ano'])), row['Inscritos'],
                 str(int(row['Inscritos'])), fontsize=8,
                 ha='center', va='bottom', color=colors(i))

    # Plot previsão
    plt.plot(pd.to_datetime(manual_forecast['Ano'], format='%Y'),
             manual_forecast['Inscritos'], marker='o', linestyle='--',
             color=colors(i), label=f'{course_name} - Previsão')

    for _, row in manual_forecast.iterrows():
        plt.text(pd.to_datetime(str(row['Ano'])), row['Inscritos'] + 5,
                 str(int(row['Inscritos'])), fontsize=8,
                 ha='center', va='bottom', color=colors(i))

# Configurações do gráfico
anos_total = sorted(df['Ano'].tolist() + [2025, 2026, 2027])
plt.xticks(pd.to_datetime([str(ano) for ano in anos_total], format='%Y'),
           [str(ano) for ano in anos_total],
           rotation=45, ha='right')

plt.xlabel('Ano')
plt.ylabel('Número de Inscritos')
plt.title('Vestibular de 2º Semestre - Gestão Empresarial - EAD (Histórico e Projeção)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()

# Geração da tabela final
final_table_df = pd.concat(all_course_data_for_table).sort_values(by=['Ano', 'NomeCurso']).reset_index(drop=True)
pivot_table_df = final_table_df.pivot_table(index='Ano', columns='NomeCurso', values='Inscritos', aggfunc='first')

print("\n--- Vestibular 2º Semestre - Inscritos por Curso (Histórico e Projeção Manual) ---")
print(pivot_table_df.to_string())
```
